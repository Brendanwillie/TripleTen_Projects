```python
#calculating recovery
C = train_data['rougher_output_concentrate_au']
F = train_data['rougher_input_feed_au']
T = train_data['rougher_output_tail_au']

calculated_recovery = (C * (F - T)) / (F * (C - T)) * 100
print(calculated_recovery)
print(train_data['rougher_output_recovery'])
```

```python
#smape function
def smape(y_true, y_pred):
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    diff = np.abs(y_true - y_pred) / denominator
    diff[denominator == 0] = 0  
    return np.mean(diff) * 100

#building first model for rougher metric

rougher_train_features = train_data_clean.iloc[:, 49:87]
rougher_train_target = train_data_clean['rougher_output_recovery']
rougher_test_features = test_data_clean.iloc[:, 49:87]



X_train, X_val, y_train, y_val = train_test_split(
    rougher_train_features, rougher_train_target, 
    test_size=0.2,  
    random_state=42)

# Initialize model
rougher_rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit model on training set
rougher_rf_model.fit(X_train, y_train)

# Predict on validation set
y_pred = rougher_rf_model.predict(X_val)

# Calculate SMAPE
smape_value_a = smape(y_val.values, y_pred)
print(f"SMAPE: {smape_value:.2f}%")

```
