## 3.4 Platform Performance Analysis

Using your selected time period, let's analyze platform performance:

````python
# Analyze platform sales trends

# Group df_relevant by platform and year_of_release to show total sales by platform per year (2009-2011)
print("Platform Sales by Year (2013-2015 Period)")
print("=" * 45)

# Group by both platform and year_of_release using df_relevant
platform_year_sales_relevant = df_relevant.groupby(['platform', 'year_of_release'])['total_sales'].sum().round(2).reset_index()

# Sort by year, then by total sales within each year
platform_year_sales_relevant = platform_year_sales_relevant.sort_values(['year_of_release', 'total_sales'], ascending=[True, False])

print("Platform Sales by Year (df_relevant data):")
display(platform_year_sales_relevant)
````

````python
# Sort platforms by total sales
platform_year_sales_relevant = df_relevant.groupby(['platform', 'year_of_release'])['total_sales'].sum().round(2).reset_index()

platform_totals_relevant = platform_year_sales_relevant.groupby('platform')['total_sales'].sum().round(2)

platform_totals_relevant = platform_totals_relevant.sort_values(ascending=False)

print("PLATFORM TOTAL SALES (2013-2015):")
print("=" * 40)
print(platform_totals_relevant)
````

````python
# Visualize top platforms
print(platform_totals_relevant.head(3))
# Calculate year-over-year growth for each platform
top_3_platforms = platform_totals_relevant.head(3)
top_platforms_list = ['X360', 'PS3', 'Wii']

# Create the growth data
growth_data = []

for platform in top_platforms_list:
    # Get sales for each year
    sales_2013 = df_relevant[(df_relevant['platform'] == platform) & 
                            (df_relevant['year_of_release'] == 2013)]['total_sales'].sum()
    sales_2014 = df_relevant[(df_relevant['platform'] == platform) & 
                            (df_relevant['year_of_release'] == 2014)]['total_sales'].sum()
    sales_2015 = df_relevant[(df_relevant['platform'] == platform) & 
                            (df_relevant['year_of_release'] == 2015)]['total_sales'].sum()
    
    # Calculate growth percentages
    growth_2013_2014 = ((sales_2014 - sales_2013) / sales_2013) * 100 if sales_2013 > 0 else 0
    growth_2014_2015 = ((sales_2015 - sales_2014) / sales_2014) * 100 if sales_2014 > 0 else 0
    
    growth_data.append({
        'Platform': platform,
        'Growth_2013_to_2014_%': round(growth_2013_2014, 1),
        'Growth_2014_to_2015_%': round(growth_2014_2015, 1)
    })
# Create the DataFrame
growth_df = pd.DataFrame(growth_data)

print("Year-over-Year Growth Percentages:")
display(growth_df)

# Your code here to calculate and visualize platform growth rates

plt.figure(figsize=(12, 8))

# Get the top 3 platforms
top_3_platforms = ['X360', 'PS3', 'Wii']
years = [2013, 2014, 2015]

# Create data for each platform
for platform in top_3_platforms:
    platform_sales = []
    for year in years:
        year_sales = df_relevant[(df_relevant['platform'] == platform) & 
                                (df_relevant['year_of_release'] == year)]['total_sales'].sum()
        platform_sales.append(year_sales)
    
    # Plot the line for this platform
    plt.plot(years, platform_sales, marker='o', linewidth=3, markersize=8, 
             label=platform, alpha=0.8)
    
    # Add value labels on each point
    for year, sales in zip(years, platform_sales):
        plt.annotate(f'{sales:.1f}M', (year, sales), 
                    textcoords="offset points", xytext=(0,10), 
                    ha='center', fontsize=10, fontweight='bold')

plt.title('Platform Sales Trends (2013-2015)\nTop 3 Platforms by Total Sales', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales (Millions)', fontsize=12)
plt.legend(fontsize=12, loc='upper right')
plt.grid(True, alpha=0.3)
plt.xticks(years)

# Set y-axis to start from 0 for better comparison
plt.ylim(0, max([df_relevant[(df_relevant['platform'] == p) & 
                            (df_relevant['year_of_release'] == y)]['total_sales'].sum() 
                for p in top_3_platforms for y in years]) * 1.1)

plt.tight_layout()
plt.show()
````

## 3.5 Sales Distribution Analysis

````python
from scipy import stats

# Set alpha threshold
alpha = 0.05
print(f"Alpha threshold: {alpha}")

# Test 1: Average user ratings of Xbox One and PC platforms are the same
print("\n" + "="*60)
print("TEST 1: Xbox One vs PC User Ratings (2013-2015)")
print("="*60)

# Null hypothesis (H0): Mean user ratings for Xbox One = Mean user ratings for PC
# Alternative hypothesis (H1): Mean user ratings for Xbox One ‚â† Mean user ratings for PC

xbox_ratings = df_relevant[df_relevant['platform'] == 'XOne']['user_score'].dropna()
pc_ratings = df_relevant[df_relevant['platform'] == 'PC']['user_score'].dropna()

print(f"Xbox One games with ratings: {len(xbox_ratings)}")
print(f"PC games with ratings: {len(pc_ratings)}")
print(f"Xbox One mean rating: {xbox_ratings.mean():.2f}")
print(f"PC mean rating: {pc_ratings.mean():.2f}")

# Perform two-sample t-test
t_stat1, p_value1 = stats.ttest_ind(xbox_ratings, pc_ratings)
print(f"\nT-statistic: {t_stat1:.4f}")
print(f"P-value: {p_value1:.4f}")

if p_value1 < alpha:
    print(f"Result: Reject H0 (p < {alpha})")
    print("Conclusion: There IS a significant difference in user ratings between Xbox One and PC")
else:
    print(f"Result: Fail to reject H0 (p >= {alpha})")
    print("Conclusion: There is NO significant difference in user ratings between Xbox One and PC")

# Test 2: Average user ratings for Action and Sports genres are different
print("\n" + "="*60)
print("TEST 2: Action vs Sports Genre User Ratings (2013-2015)")
print("="*60)

# Null hypothesis (H0): Mean user ratings for Action = Mean user ratings for Sports
# Alternative hypothesis (H1): Mean user ratings for Action ‚â† Mean user ratings for Sports

action_ratings = df_relevant[df_relevant['genre'] == 'Action']['user_score'].dropna()
sports_ratings = df_relevant[df_relevant['genre'] == 'Sports']['user_score'].dropna()

print(f"Action games with ratings: {len(action_ratings)}")
print(f"Sports games with ratings: {len(sports_ratings)}")
print(f"Action mean rating: {action_ratings.mean():.2f}")
print(f"Sports mean rating: {sports_ratings.mean():.2f}")

# Perform two-sample t-test
t_stat2, p_value2 = stats.ttest_ind(action_ratings, sports_ratings)
print(f"\nT-statistic: {t_stat2:.4f}")
print(f"P-value: {p_value2:.4f}")

if p_value2 < alpha:
    print(f"Result: Reject H0 (p < {alpha})")
    print("Conclusion: There IS a significant difference in user ratings between Action and Sports genres")
else:
    print(f"Result: Fail to reject H0 (p >= {alpha})")
    print("Conclusion: There is NO significant difference in user ratings between Action and Sports genres")

print("\n" + "="*60)
print("METHODOLOGY & EXPLANATION")
print("="*60)
print("Hypothesis Testing Framework:")
print("- H0 (null): The means are equal (no difference)")
print("- H1 (alternative): The means are not equal (there is a difference)")
print("\nStatistical Method:")
print("- Two-sample t-test: Compares means of two independent groups")
print("- Chosen because we have continuous data (ratings) and want to compare group means")
print("- Assumes normal distribution and independent samples")
print("\nSignificance Level:")
print("- Alpha = 0.05 (5%): Standard threshold in social sciences")
print("- Balances Type I error risk with statistical power")
print("- If p-value < 0.05, reject H0 (significant difference exists)")
print("- If p-value >= 0.05, fail to reject H0 (no significant difference)")
````
