# Project description

Sweet Lift Taxi company has collected historical data on taxi orders at airports. To attract more drivers during peak hours, we need to predict the amount of taxi orders for the next hour. Build a model for such a prediction.

The RMSE metric on the test set should not be more than 48.

## Project instructions

1. Download the data and resample it by one hour.
2. Analyze the data.
3. Train different models with different hyperparameters. The test sample should be 10% of the initial dataset. 
4. Test the data using the test sample and provide a conclusion.

## Data description

The data is stored in file `taxi.csv`. The number of orders is in the '*num_orders*' column.

## Preparation

````python
# imports/veiw dataset
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from sklearn.model_selection import train_test_split
import pmdarima as pm
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
from statsmodels.tsa.stattools import arma_order_select_ic
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor



df = pd.read_csv('/datasets/taxi.csv', index_col=[0], parse_dates=[0])
model_data = df['2018-03' : '2018-08'].resample('1H').sum()
display(df.head())
print(df.info)
````

## Analysis

````python
# seasonality across the whole set
data_subset= df['2018-03' : '2018-08'].resample('1D').sum()
decomposed = seasonal_decompose(data_subset)

plt.figure(figsize=(6, 8))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend')
plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality')
plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals')
plt.tight_layout()
````

````python
#seasonality across 2 weeks
data_subset_H= df['2018-03-01' : '2018-03-15'].resample('1H').sum()
decomposed = seasonal_decompose(data_subset_H)

plt.figure(figsize=(6, 8))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend')
plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality')
plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals')
plt.tight_layout()
````

````python
#stationary testing/resampling model data

#full hourly original dataset
df_stationaryTest = adfuller(df, autolag= 'AIC')
print('P-value: ', df_stationaryTest[1])

#resampled to one day dataset to reduce noise
df_stationaryTest = adfuller(data_subset_H, autolag= 'AIC')
print('P-value: ', df_stationaryTest[1])

#our hourly model dataset
df_stationaryTest = adfuller(model_data, autolag= 'AIC')
print('P-value: ', df_stationaryTest[1])
````

## Training

````python
#splitting the data

features = model_data.drop('num_orders', axis=1)  
target = model_data['num_orders'] 

clean_data = model_data.dropna()
features = clean_data.drop('num_orders', axis=1)
target = clean_data['num_orders']

features_temp, features_test, target_temp, target_test = train_test_split(
    features, target,
    test_size=0.10,
    shuffle=False,
    random_state=42)

features_train, features_val, target_train, target_val = train_test_split(
    features_temp, target_temp,
    test_size=0.20,
    shuffle=False,
    random_state=42)

print("Training set size:", len(features_train))
print("Validation set size:", len(features_val))
print("Test set size:", len(features_test))
````

````python
arima_order = pm.auto_arima(target_train, seasonal=True, m=6, max_p=3, max_q=3, stepwise=True, suppress_warnings=True)
print("ARIMA order:", arima_order.order)
print("Seasonal order:", arima_order.seasonal_order)
````

````python
arma_order = pm.auto_arima(
    target_train, 
    seasonal=True,
    m=6,             
    d=0,             
    D=0,             
    max_d=0,         
    max_D=0,         
    max_p=3, 
    max_q=3,
    stepwise=True, 
    suppress_warnings=True
)
print("ARMA order:", arma_order.order)
````

````python
# Fit both models on train data
arima_model = ARIMA(target_train, order=arima_order.order, seasonal_order=arima_order.seasonal_order)
arima_result = arima_model.fit()

arma_model = ARIMA(target_train, order=arma_order.order)  
arma_result = arma_model.fit()

# Predictions 
arima_train_pred = arima_result.fittedvalues
arma_train_pred = arma_result.fittedvalues

arima_val_pred = arima_result.forecast(steps=len(target_val))
arma_val_pred = arma_result.forecast(steps=len(target_val))

# Calculate Train RMSE
arima_train_rmse = np.sqrt(mean_squared_error(target_train, arima_train_pred))
arma_train_rmse = np.sqrt(mean_squared_error(target_train, arma_train_pred))

# Calculate Validation RMSE
arima_val_rmse = np.sqrt(mean_squared_error(target_val, arima_val_pred))
arma_val_rmse = np.sqrt(mean_squared_error(target_val, arma_val_pred))

# Print Results
print("Training RMSE:")
print(f"ARIMA: {arima_train_rmse:.2f}")
print(f"ARMA: {arma_train_rmse:.2f}")

print("\nValidation RMSE:")
print(f"ARIMA: {arima_val_rmse:.2f}")
print(f"ARMA: {arma_val_rmse:.2f}")
````

ARMA and ARIMA just arn quite cutting it, there could potentially be some over fitting with these mroe complex models. We are gonna try something simpler.

````python
# MA model
res = arma_order_select_ic(y=target_train, max_ar=0, max_ma=10)
ma_order = res.bic_min_order[1]
print(f"Selected MA order: {ma_order}")

ma_model = ARIMA(target_train, order=(0, 0, ma_order))
ma_model_fitted = ma_model.fit()

start_value = len(target_train)
end_value = len(target_train) + len(target_val) - 1
ma_val_pred = ma_model_fitted.predict(start=start_value, end=end_value, dynamic=False)

ma_train_pred = ma_model_fitted.fittedvalues

ma_train_rmse = np.sqrt(mean_squared_error(target_train, ma_train_pred))
ma_val_rmse = np.sqrt(mean_squared_error(target_val, ma_val_pred))

print(f"MA Train RMSE: {ma_train_rmse:.2f}")
print(f"MA Validation RMSE: {ma_val_rmse:.2f}")
````

````python
#some manual adjusting because its still overfitting
ma_model = ARIMA(target_train, order=(0, 0, 1))
ma_model_fitted = ma_model.fit()

start_value = len(target_train)
end_value = len(target_train) + len(target_val) - 1
ma_val_pred = ma_model_fitted.predict(start=start_value, end=end_value, dynamic=False)

ma_train_pred = ma_model_fitted.fittedvalues

ma_train_rmse = np.sqrt(mean_squared_error(target_train, ma_train_pred))
ma_val_rmse = np.sqrt(mean_squared_error(target_val, ma_val_pred))

print(f"MA Train RMSE: {ma_train_rmse:.2f}")
print(f"MA Validation RMSE: {ma_val_rmse:.2f}")
````

````python
#linear regression
max_lag = 6
def make_features(data, max_lag, rolling_mean_size):
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour

    for lag in range(1, max_lag + 1):
        data[f'lag_{lag}'] = data['num_orders'].shift(lag)

    data['rolling_mean'] = (
        data['num_orders'].shift().rolling(rolling_mean_size).mean()
    )

model_data_fe = model_data.copy()
make_features(model_data_fe, 6, rolling_mean_size=10)

model_data_clean = model_data_fe.dropna()

train_features = model_data_clean.loc[model_data_clean.index.intersection(features_train.index)]
val_features = model_data_clean.loc[model_data_clean.index.intersection(features_val.index)]


feature_columns = ['year', 'month', 'day', 'dayofweek', 'hour', 'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5', 'lag_6', 'rolling_mean']

X_train = train_features[feature_columns]
y_train = train_features['num_orders']
X_val = val_features[feature_columns]
y_val = val_features['num_orders']

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

lr_train_pred = lr_model.predict(X_train)
lr_val_pred = lr_model.predict(X_val)

lr_train_rmse = np.sqrt(mean_squared_error(y_train, lr_train_pred))
lr_val_rmse = np.sqrt(mean_squared_error(y_val, lr_val_pred))

print(f"Linear Regression Train RMSE: {lr_train_rmse:.2f}")
print(f"Linear Regression Validation RMSE: {lr_val_rmse:.2f}")
````

````python
#linear regression hypertuning
best_rmse = float('inf')
best_params = {}
results = []

max_lag_options = [3, 4, 5, 6, 7, 8]
rolling_window_options = [5, 7, 10, 12, 15, 20]


model_data_fe = model_data.copy()

for max_lag in max_lag_options:
    for rolling_window in rolling_window_options:
        make_features(model_data_fe, max_lag=max_lag, rolling_mean_size=rolling_window)
        train_temp = model_data_fe.loc[features_train.index].dropna()
        val_temp = model_data_fe.loc[features_val.index].dropna()
        
        feature_columns = (
            ['year', 'month', 'day', 'dayofweek', 'hour'] +
            [f'lag_{i}' for i in range(1, max_lag + 1)] +
            ['rolling_mean']
        )
        
        X_train = train_temp[feature_columns]
        y_train = train_temp['num_orders']
        X_val = val_temp[feature_columns]
        y_val = val_temp['num_orders']
        
        lr_model = LinearRegression()
        lr_model.fit(X_train, y_train)
        
        val_pred = lr_model.predict(X_val)
        rmse = np.sqrt(mean_squared_error(y_val, val_pred))
        
        results.append((max_lag, rolling_window, rmse))
        
        if rmse < best_rmse:
            best_rmse = rmse
            best_params = (max_lag, rolling_window)
            print(f"New best RMSE: {best_rmse:.3f} | "
                  f"max_lag={max_lag}, rolling_mean_size={rolling_window}")

print("\n=== BEST PARAMETERS FOUND ===")
print(f"Best RMSE: {best_rmse:.3f}")
print(f"Best max_lag: {best_params[0]}")
print(f"Best rolling_mean_size: {best_params[1]}")
````

<b>‚ö†Ô∏è Reviewer's comment V1</b> 
    
Lag and rolling window are features, not hyperparameters. This is feature engineering, and is a valid way to improve performance of a model, however it is not hyperparameter tuning.

````python
# AR model
res = arma_order_select_ic(y=target_train, max_ar=10, max_ma=0)
ar_order = res.bic_min_order[0]
print(f"Selected AR order: {ar_order}")

ar_model = ARIMA(target_train, order=(ar_order, 0, 0))
ar_model_fitted = ar_model.fit()

start_value = len(target_train)
end_value = len(target_train) + len(target_val) - 1
ar_val_pred = ar_model_fitted.predict(start=start_value, end=end_value, dynamic=False)

ar_train_pred = ar_model_fitted.fittedvalues

ar_train_rmse = np.sqrt(mean_squared_error(target_train, ar_train_pred))
ar_val_rmse = np.sqrt(mean_squared_error(target_val, ar_val_pred))

print(f"AR Train RMSE: {ar_train_rmse:.2f}")
print(f"AR Validation RMSE: {ar_val_rmse:.2f}")
````

````python
ar_model = ARIMA(target_train, order=(1, 0, 0))
ar_model_fitted = ar_model.fit()

start_value = len(target_train)
end_value = len(target_train) + len(target_val) - 1
ar_val_pred = ar_model_fitted.predict(start=start_value, end=end_value, dynamic=False)

ar_train_pred = ar_model_fitted.fittedvalues

ar_train_rmse = np.sqrt(mean_squared_error(target_train, ar_train_pred))
ar_val_rmse = np.sqrt(mean_squared_error(target_val, ar_val_pred))

print(f"AR Train RMSE: {ar_train_rmse:.2f}")
print(f"AR Validation RMSE: {ar_val_rmse:.2f}")
````

````python
# Random Forest hyper tuning
best_rmse = float('inf')
best_params = {}

for n_estimators in [50, 100, 200]:
    for max_depth in [5, 10, 15]:
        # Fit and train the model with current hyperparameters
        model = RandomForestRegressor(
            n_estimators=n_estimators,
            max_depth=max_depth,
            random_state=42
        )
        model.fit(X_train, y_train)
        
        # Make predictions and evaluate
        val_pred = model.predict(X_val)
        rmse = np.sqrt(mean_squared_error(y_val, val_pred))
        
        # Track best performance
        if rmse < best_rmse:
            best_rmse = rmse
            best_params = {'n_estimators': n_estimators, 'max_depth': max_depth}
            
print(f"Best RMSE: {best_rmse}")
print(f"Best parameters: {best_params}")
````

````python
#building further with more features/more focused lag selection on linear regression
def make_enhanced_features(data, rolling_mean_size):
    lag_list = [24, 48, 168]
    
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    data['is_weekend'] = (data.index.dayofweek >= 5).astype(int)
    data['is_morning_rush'] = ((data.index.hour >= 7) & (data.index.hour <= 9)).astype(int)
    data['is_evening_rush'] = ((data.index.hour >= 17) & (data.index.hour <= 19)).astype(int)
    data['hour_weekend_interaction'] = data['hour'] * data['is_weekend']
    
    for lag in lag_list:
        data[f'lag_{lag}'] = data['num_orders'].shift(lag)
    
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()

model_data_fe = model_data.copy()
make_enhanced_features(model_data_fe, rolling_mean_size=5)

model_data_clean = model_data_fe.dropna()

train_features = model_data_clean.loc[model_data_clean.index.intersection(features_train.index)]
val_features = model_data_clean.loc[model_data_clean.index.intersection(features_val.index)]

feature_columns = [
    'year', 'month', 'day', 'dayofweek', 'hour',
    'lag_24', 'lag_48', 'lag_168', 'rolling_mean',
    'is_weekend', 'is_morning_rush', 'is_evening_rush', 'hour_weekend_interaction'
]

X_train = train_features[feature_columns]
y_train = train_features['num_orders']

X_val = val_features[feature_columns]
y_val = val_features['num_orders']

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

lr_train_pred = lr_model.predict(X_train)
lr_val_pred = lr_model.predict(X_val)

lr_train_rmse = np.sqrt(mean_squared_error(y_train, lr_train_pred))
lr_val_rmse = np.sqrt(mean_squared_error(y_val, lr_val_pred))

print(f"Linear Regression Train RMSE: {lr_train_rmse:.2f}")
print(f"Linear Regression Validation RMSE: {lr_val_rmse:.2f}")
````
